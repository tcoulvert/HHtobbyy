```python
# ML packages
from torch.utils.data import DataLoader
from lightning import Trainer
from lightning.pytorch.callbacks.early_stopping import EarlyStopping

# Callbacks
callbacks = [EarlyStopping(monitor=model.modelconfig.monitor, min_delta=model.modelconfig.min_delta, patience=model.modelconfig.patience, verbose=False, mode=model.modelconfig.mode)]

trainer = Trainer(
    callbacks=callbacks,
    default_root_dir=model.modelconfig.output_dirpath,
    max_epochs=model.modelconfig.max_epochs, 
    accelerator=model.modelconfig.accelerator,
    strategy=model.modelconfig.strategy,
    num_nodes=model.modelconfig.num_nodes,
    precision=model.modelconfig.precision, 
    gradient_clip_val=model.modelconfig.gradient_clip_val,
    logger=model.modelconfig.logger,
    devices=1 if eval else model.modelconfig.devices,
)

mlp_model.load_state_dict(pth['model_state_dict'])
trainer.predict(mlp_model, DataLoader([]))
trainer.save_checkpoint("manos_trained_model.ckpt")

mlp_model.load_state_dict(pth['best_weights'])
trainer.predict(mlp_model, DataLoader([]))
trainer.save_checkpoint("manos_best_model.ckpt")
```
